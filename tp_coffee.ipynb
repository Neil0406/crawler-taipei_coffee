```python
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
import requests
res = requests.get('https://cafenomad.tw/taipei/list')
soup = bs(res.text,'lxml')
```

```python
links = []
name = []
quiet = []
mrt = []
optime = []


data = soup.find_all('a','seo-link')
for i in data:
    links.append(f'https://cafenomad.tw/{i.get("href")}')
    name.append(i.text)
                 
data = soup.select('td.c3.-small')   #quiet
for i in data:
    quiet.append(i.text)
    
r = []    
data = soup.select('td.c9.-small')             #mrt
for i in data:
    r.append(i.text.split())
for i in r:
    try:
        mrt.append(i[0])
    except:
        mrt.append('')
r2 = []
w = ''
data = soup.select('td.c11.-medium._ellipsis')   #optime        
for i in data:
    r2.append(i.text.split())
for i in r2:
    try:
        optime.append(w.join(i))
    except:
        optime.append('')

        
city = []
area = []
wifi = []
sit = []
time = []
socket = []
map_links = []
location = []
                 
for i in tqdm(links):
    res = requests.get(i)
    soup = bs(res.text,'lxml')

    w = ''
    ar = soup.select(f'#modal-{i[27:]} > div:nth-child(2)')
    a = ar[0].text.split()
    try:
        city.append(a[0])
    except:
        city.append('') 
    try:
        area.append(a[2])
    except:
        area.append('')        
#------------------------------   
    wi = soup.select('div:nth-child(8) > div:nth-child(1) > div:nth-child(1) > div')      # wifi
    if wi == []:
        wi = soup.select('div:nth-child(7) > div:nth-child(1) > div:nth-child(1) > div')
               
    try:
        wifi.append(wi[0].text) 
    except:
        wifi.append('') 
#------------------------------
    si = soup.select('div:nth-child(8) > div:nth-child(1) > div:nth-child(2) > div') #sit
    if si == []:
        si = soup.select('div:nth-child(7) > div:nth-child(1) > div:nth-child(2) > div')
               
    try:
        sit.append(si[0].text)
    except:
        sit.append('')
#------------------------------  
    ti = soup.select('div:nth-child(14) > div:nth-child(3) > div:nth-child(1) > div')
    if ti == []:
        ti = soup.select('div:nth-child(13) > div:nth-child(3) > div:nth-child(1) > div')

    try:
        time.append(ti[0].text)
    except:
        time.append('')
#------------------------------                 
    so = soup.select('div:nth-child(14) > div:nth-child(3) > div:nth-child(2) > div')
    if so == []:
        so = soup.select('div:nth-child(13) > div:nth-child(3) > div:nth-child(2) > div')

    try:
        socket.append(so[0].text)
    except:
        socket.append('')
#------------------------------
    data = soup.select('div:nth-child(16) > div:nth-child(2) > div:nth-child(3) > a')      #links
    if data == []:
        data = soup.select('div:nth-child(15) > div:nth-child(2) > div:nth-child(3) > a')   

    try:
        map_links.append(data[0].get('href'))
    except:
        map_links.append('')
    try:
        location.append(data[0].text.replace('\n',''))
    except:
        location.append('')


```
